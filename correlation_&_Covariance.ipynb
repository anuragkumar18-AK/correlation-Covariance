##**QUESTION -> 1**   

##Define Covariance and explain how it differs from Correlation in terms of scale and interpretation.

##***Answer -->***

* Covariance: Measures the direction of the linear relationship between two variables. If both variables increase together, covariance is positive; if one increases while the other decreases, it is negative.  

* Correlation: A standardized version of covariance that measures both the strength and direction of the linear relationship.  



* **```Key Differences:```**

* Scale: Covariance is expressed in the units of the variables (e.g., $cm \times kg$), making it hard to compare. Correlation is dimensionless, ranging strictly from $-1$ to $+1$.

* Interpretation: Correlation is preferred because its scale-invariance allows you to determine how "strong" a relationship is regardless of the units used.

##**QUESTION -> 3**

##iscuss the limitations of covariance as a measure of relationship between two variables. Why is correlation preferred in many cases?

##***Answer -->***

 **```Why is correlation preferred in many cases?```**   

While covariance indicates the direction of a relationship, it has several mathematical and practical limitations:

* ***Sensitivity to Scale:*** The value of covariance depends entirely on the units of the variables. If you change the measurement from meters to centimeters, the covariance value will increase by 100, even though the relationship between the variables hasn't changed.  

* ***Lack of Strength Assessment:*** Covariance does not have a set range (it can be any number from $-\infty$ to $+\infty$). This makes it impossible to determine how "strong" the relationship is just by looking at the number.  

* ***Difficulty in Comparison:*** You cannot easily compare the covariance of "Height vs. Weight" with "Income vs. Education" because they use different units.

**```Why Correlation is Preferred?```**

Correlation is essentially a "standardized" version of covariance that solves these issues:

* **Fixed Range:** Correlation is always between $-1$ and $+1$.  

* **Unitless:** It is a pure number, meaning it is not affected by the scale of measurement.  

* **Interpretability:** Because of the fixed range, a value of $0.85$ immediately tells you the relationship is very strong and positive, whereas a covariance of $0.85$ tells you very little without knowing the units.

##**QUESTION -> 4**

##Explain the difference between Pearson’s correlation coefficient and Spearman’s rank correlation coefficient. When would you prefer to use Spearman’s correlation?

##***Answer -->***

* Pearson’s ($r$): Measures the strength of a linear relationship between continuous variables.  

* Spearman’s ($\rho$): Measures the monotonic relationship using the ranks of the data.  

* When to use Spearman: Use it when the data is ordinal (ranked), non-linearly related, or contains significant outliers that would skew Pearson’s results.
